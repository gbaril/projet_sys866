```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def getFile(path, maml):
    return 'logs/{}/{}.csv'.format('maml' if maml else 'proto_nets', path)

df = pd.read_csv(getFile('omniglot_order=1_n=5_k=20_metabatch=16_train_steps=5_val_steps=5', True), sep=',')

x = list(range(df.shape[0]))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))

ax = axes[0]

#ax = plt.axes()
ax.plot(x, df['categorical_accuracy'], label='On validation set')
ax.plot(x, df[df.columns[-2]], label='On new classes')
ax.set_title('Accuracy at each epochs')
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.legend()

ax = axes[1]

#ax = plt.axes()
ax.plot(x, df['loss'], label='Training Loss')
ax.plot(x, df['val_loss'], label='Validation Loss')
ax.set_title('Loss at each epochs')
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.legend()
plt.show()
```
